```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, output_file
from bokeh.models import ColumnDataSource, mappers
output_notebook()
from PIL import Image
from bokeh.palettes import viridis

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/final.csv")

bean = np.array(df["bean"]).reshape(500, 500).astype("uint32")
bean = np.flipud(bean)
output_file("test.html")

mapper = mappers.LinearColorMapper(
    palette="Viridis256",
    low = bean.min(),
    high = bean.max()
)

p = figure(width = 600, height = 600, tooltips=[("(col,row)", "($x, $y)"), ("value", "@image")])
p.image(image=[bean], x=0, y=0, dw=500, dh=500, color_mapper = mapper)
show(p)
```
